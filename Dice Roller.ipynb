```python
import random
import matplotlib.pyplot as plt
import time

start_time = time.time()

def roll_dice(sides, number):
    return [random.randint(1, sides) for _ in range(number)]

def max_roll(sides, number, times):
    max_values = []
    for _ in range(times):
        rolls = roll_dice(sides, number)
        max_value = max(rolls)
        max_values.append(max_value)
    return max_values

sides = int(input("Enter the number of sides on the dice: "))
number = int(input("Enter the number of dice to roll: "))
times = int(input("Enter the total number of times to roll: "))

results = max_roll(sides, number, times)

fig = plt.figure()
fig.suptitle(f"Max value of {number} {sides}-sided dice rolled {times} times")
plt.title("aka don't blame your dice")

plt.hist(results, bins=range(1, sides+2), align='left', rwidth=0.8, color='red')
#plt.title(f"Max value of {number} {sides}-sided dice rolled {times} times")
plt.xlabel("Max value")
plt.ylabel("Frequency")
plt.xticks(range(1, sides+1))
plt.xlim([0.5, sides+0.5])
plt.show()

time.sleep(2)

end_time = time.time()
elapsed_time = end_time - start_time

print(f"Elapsed time: {elapsed_time:.2f} seconds")
```
